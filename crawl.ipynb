```python
from icrawler.builtin import BingImageCrawler
import os
```

Crawl Car images

```python
categories = ['MPV', 'SUV', 'Sedan', 'Hatchback', 'Pickup', 'Truck', 'Bus']
```

```python
categories = {
    "MPV": ["Toyota Avanza", "Daihatsu Xenia", "Mitsubishi Xpander", 'Honda BRV'],
    "SUV": ["Toyota Rush", "Honda CR-V", "Hyundai Creta", 'Hyundai Santa FE'],
    "Sedan": ["Toyota Vios", 'Toyota Camry', 'Honda City', 'Honda Civic'],
    "Hatchback": ["Toyota Yaris", "Honda Jazz", "Honda Civic hatchback", 'Mazda 3 Hatchback'],
    "PickUp": ["Mitsubishi L300 Pick Up", "Grand Max Pick-up"],
    "Truck": ["Mitsubishi Colt diesel truk", 'Mitsubishi fuso', 'Truk container'],
    "Bus": ['Bus double decker indonesia', 'Jetbus 5', 'Jetbus HD']
}
```

```python
for category, models in categories.items():
    for item in models:
        save_path = os.path.join("data_crawl", category)
        os.makedirs(save_path, exist_ok=True)

        crawler = BingImageCrawler(storage = {"root_dir": save_path})
        crawler.crawl(keyword=f'{item}', max_num=200)
```
