```python
import os
import yaml
from ultralytics import YOLO
```

```python
# load pre-trained model from ultralytics 
model = YOLO('yolo11n.pt')
```

```python
# define data path
data_path = "./data"

# define yaml path
yaml_path = os.path.join(data_path, "data.yaml")

# Set paths for training and validation image sets
train_images_path = os.path.join(data_path, 'train', 'images')
valid_images_path = os.path.join(data_path, 'valid', 'images')
```

training

```python
model_train = model.train(
    data = yaml_path,
    epochs = 100,
    patience = 10,
    batch = 16,
    imgsz=640
)
```
